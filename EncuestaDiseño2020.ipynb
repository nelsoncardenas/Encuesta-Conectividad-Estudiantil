```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
from google.colab import files
rcParams['figure.figsize'] = [6.0, 4.0] # valores originales

sns.set()
```

```
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/EncuestaDiseño2020/EncuestaDiseño2020_IdentidadProtegida.csv')
#df.columns = ['Fecha', 'Correo', 'Nombre', 'Apellidos', 'Teléfono', 'Grado', 'Computador', 'Internet']
df.columns = ['Fecha', 'Grado', 'Computador', 'Internet']
df.head()
```

```
print('Encuesta diligenciada por:', sum(df['Grado'] == '10-5') + sum(df['Grado'] == '11-4'), 'estudiantes.')
print(sum(df['Grado'] == '10-5'), 'estudiantes de 10-5 y por', sum(df['Grado'] == '11-4'), 'estudiantes de 11-4.')
```

# **Disponibilidad de computadores en casa**

```
rcParams['figure.figsize'] = [12.0, 4.0]
plt.subplot(1, 6, 1)
df.Computador.value_counts().plot(kind='bar', rot =0)
plt.title('¿Cuenta con computador de mesa o portátil en su casa disponible para trabajar?\n GRADOS 10 y 11')
plt.xlabel('Frecuencia')

plt.subplot(1, 6, 2)
conteo_comp = df['Computador'].value_counts()
plt2 = conteo_comp.plot(kind='pie')
plt.axis('equal')

plt.subplot(1, 6, 3)
df[df['Grado']=='10-5'].Computador.value_counts().plot(kind='bar', rot =0)
plt.title('\n GRADO 10 ')
plt.xlabel('Frecuencia')

plt.subplot(1, 6, 4)
conteo_comp = df[df['Grado']=='10-5']['Computador'].value_counts()
plt2 = conteo_comp.plot(kind='pie')
plt.axis('equal')


plt.subplot(1, 6, 5)
df[df['Grado']=='11-4'].Computador.value_counts().plot(kind='bar', rot =0)
plt.title('\n GRADO 11 ')
plt.xlabel('Frecuencia')

plt.subplot(1, 6, 6)
conteo_comp = df[df['Grado']=='11-4']['Computador'].value_counts()
plt2 = conteo_comp.plot(kind='pie')
plt.axis('equal')
_=0

print('El', str(conteo_comp['Sí']/sum(conteo_comp)*100)[:5], '% de los estudiantes posee computador')
print('El', str(conteo_comp['No']/sum(conteo_comp)*100)[:5], '% no posee computador.')

fig = plt.gcf()
fig.savefig('G1.svg')
files.download("G1.svg") 
```

# **Disponibilidad de Internet en casa**

```
rcParams['figure.figsize'] = [12.0, 4.0]
plt.subplot(1, 6, 1)
df.Internet.value_counts().plot(kind='bar', rot =0)
plt.title('¿Cuenta con Internet de hogar?\n GRADOS 10 y 11')
plt.xlabel('Frecuencia')

plt.subplot(1, 6, 2)
conteo_comp = df['Internet'].value_counts()
plt2 = conteo_comp.plot(kind='pie')
plt.axis('equal')

plt.subplot(1, 6, 3)
df[df['Grado']=='10-5'].Internet.value_counts().plot(kind='bar', rot =0)
plt.title('\n GRADO 10 ')
plt.xlabel('Frecuencia')

plt.subplot(1, 6, 4)
conteo_comp = df[df['Grado']=='10-5']['Internet'].value_counts()

plt2 = conteo_comp.plot(kind='pie')
plt.axis('equal')


plt.subplot(1, 6, 5)
df[df['Grado']=='11-4'].Internet.value_counts().plot(kind='bar', rot =0)
plt.title('\n GRADO 11 ')
plt.xlabel('Frecuencia')

plt.subplot(1, 6, 6)
conteo_comp = df[df['Grado']=='11-4']['Internet'].value_counts()
plt2 = conteo_comp.astype(int).plot(kind='pie')
plt.axis('equal')
_=0

print('El', str(conteo_comp['Sí']/sum(conteo_comp)*100)[:5], '% de los estudiantes posee Internet')
print('El', str(conteo_comp['No']/sum(conteo_comp)*100)[:5], '% no posee Internet')

fig = plt.gcf()
fig.savefig('G2.svg')
files.download("G2.svg")
```
